```python
import pandas as pd
import numpy as np
```

```python
af = pd.read_csv('airfoil_self_noise.dat', sep='\t', header=None)
```

```python
af.head(10)
```

```python
af.columns = ['Frequency', 'Angle_of_attack', 'Chord_length', 'velocity', 'thickness', 'Noise']
af.head()
```

```python
af.shape
```

```python
X = af.drop('Noise', axis = 1)
y = af['Noise']
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

#### Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
lc = LinearRegression()
```

```python
lc.fit(X_train, y_train)
```

```python
lc.score(X_train, y_train)
```

```python
y_preds = lc.predict(X_test)
```

```python
y_preds
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
y_preds = lc.predict(X_test)
mae = mean_absolute_error(y_test, y_preds)
mae
```

```python
y_preds = lc.predict(X_test)
mse = print(mean_squared_error(y_test, y_preds))
mse
```

#### Using Logistic Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_train,y_train)
```

```python
y_preds = rfr.predict(X_test)
```

```python
y_preds
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, y_preds)
mae
```

```python
mse = mean_squared_error(y_test, y_preds)
mse
```

```python
np.sqrt(mse)
```
